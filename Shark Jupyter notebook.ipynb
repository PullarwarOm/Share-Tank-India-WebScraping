```python
tbody = []
for r in soup.select('table')[1].tbody.find_all('tr'):
    tbody.append([i.text.strip() for i in r.find_all('td')])
```

```python
tbody
```

```python
import pandas as pd
```

```python
def gather_table(soup):
    table1head = [i.text.strip() for i in soup.select('thead')[0].select('th')]
    tbody = []
    for r in soup.select('table')[1].tbody.find_all('tr'):
        tbody.append([i.text.strip() for i in r.find_all('td')])
    tb1=pd.DataFrame(tbody, columns = table1head)
    return(tb1)
```

```python
gather_table(soup)
```

```python
gather_table(soup).to_csv(r'C:\Users\Admin\Desktop\Assignment\Data Scraping\Assignmetn solution\Question 1/Shark Tank India Investors.csv', index=False)
```
